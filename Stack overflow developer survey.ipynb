```python
def shorten_age(categories,cut_off):
    categorical_map={}
    for i in range(len(categories)):
        if categories.values[i] >=cut_off:
             categorical_map[categories.index[i]]=categories.index[i]
        else:
            categorical_map[categories.index[i]]='Other_age'
    return categorical_map    
```

## Now Clean the Education level column :

```python
def clean_education(x):
    if 'Master’s degree ' in x:
        return 'Masters'
    if 'Bachelor’s degree' in x:
        return "Bachelaors"
    if 'Other doctoral degree' in x :
        return 'Post Grad'
    return 'Less than Bachelors'
```

```python
df['EdLevel']=df['EdLevel'].apply(clean_education)
```

## now apply Grid Search CV:

```python
from sklearn.model_selection import GridSearchCV
```

```python
max_depth=[None,2,4,6,8,12]
parameters={'max_depth':max_depth}

rg=DecisionTreeRegressor(criterion='squared_error',min_samples_split=2,random_state=0)
gs=GridSearchCV(rg,parameters,scoring='neg_mean_squared_error')
gs.fit(X,y)
```

```python
gs.best_estimator_
```

```python
regressor=gs.best_estimator_
```

```python
regressor.fit(X,y)
```

```python
y_pred4=regressor.predict(X)
```

```python
error4=np.sqrt(mean_squared_error(y,y_pred4))
print('${: ,.02f}'.format(error4))
```
